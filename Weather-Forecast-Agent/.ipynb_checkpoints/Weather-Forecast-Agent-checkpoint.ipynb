```python
import requests

BASE_URL = "https://api.openweathermap.org/data/2.5/weather"
API_KEY = "0d11fed55926efd2983324b73cf8f37c"


#custom weather forecast function

def find_weather(city: str) -> dict:
    """This function returns the current weather forecast for the given city"""
    units = "metric"
    params = {
        'q': city,
        'appid': API_KEY,
        'units': units
    }
    
    response = requests.get(BASE_URL, params=params)
    result = response.json()
    return result
```

```python
import os
import requests
from pydantic import BaseModel
from pydantic_ai import Agent, RunContext
from pydantic_ai.settings import ModelSettings
```

```python
os.environ["GROQ_API_KEY"] = "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"
```

```python
#Define the output schema
class WeatherForecast(BaseModel):
    location: str
    description: str
    temperature_celsius: float
```

```python
weather_agent = Agent(
    model="groq:llama-3.1-8b-instant",
    model_settings=ModelSettings(temperature=0.2),
    output_type=str,
    system_prompt="You are a helpful assitent. Use the 'get_weather_forecast' tool to find current weather conditions for any city. Provide clean and friendly answer"
)
```

```python
#Weather forecast tool
@weather_agent.tool
def get_weather_forecast(ctx: RunContext, city) -> WeatherForecast:
    """
    Return current weather forecast for a city using OpenWeatherMap API
    """
    
    url = "https://api.openweathermap.org/data/2.5/weather"
    api_key = "0d11fed55926efd2983324b73cf8f37c"

    params = {
        'q': city,
        'appid': api_key,
        'units': 'metric'
    }
    
    res = requests.get(url, params=params).json()
    return WeatherForecast(
       location=res["name"],
       description = res["weather"][0]["description"].capitalize(),
       temperature_celsius = res["main"]["temp"]
        
    )
```

```python
question = input(" Ask about the weather: ")
result = weather_agent.run_sync(question)
print("\n Forecast: ", result.output)
```
